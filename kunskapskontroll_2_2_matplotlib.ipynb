### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

There are two ways of plotting in Matplotlib:

1. Pyplot approach or (Implicit gränssnitt):
    Payplot approach uses functions like plt.plot(), plt.scatter(). Its simple and convenient for quick plotting.
    Example: plt.plot(x, y)

2. Object-Oriented approach or (Explicit gränssnitt):
    Object-oriented approach is more flexible and recommended for complex plots. We can explicitly create Figure and Axes objects,
    then call methods on the Axes.
    Example: fig, ax = plt.subplots()
             ax.plot(x, y)

Object-Oriented approach is recommended approach because it is clearer, more organized, and works better with multiple subplots.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: An image that may contain axes   
Axes: Represents a single plot with the figure   
Axis: The x-axis, y-axis or z-axis of a plot   
Artist: Every thing visible on the figure is an artist like lines, text, shapes and legends.

### When plotting in Matplotlib, what is the expected input data type?

When plotting in Matplotlib, expected input data types are usually NumPy arrays (np.array) or Python lists.   
And x and y values should have the same length.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100)
y = x**2

# By using Object-Oriented approach
fig, ax = plt.subplots()
ax.plot(x, y, label='y = x^2')
ax.set_title("Plot of function y = x^2")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.show()
```

```python
# By using pyplot approach
plt.plot(x, y, color='red', label='y = x^2')
plt.title("Plot of y = x^2 by pyplot approach")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
# Create figure with 2 subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Scatter plot
axes[0].scatter(random_data_x, random_data_y, color='purple', alpha=0.6)
axes[0].set_title("Scatter Plot")
axes[0].set_xlabel("X values")
axes[0].set_ylabel("Y values")

# Bar plot
axes[1].bar(names, values, color='green', )
axes[1].set_title("Bar plot for Fruit Counts")
axes[1].set_xlabel("Fruit")
axes[1].set_ylabel("Count")

plt.tight_layout()  # Adjust spacing
plt.show()
```
